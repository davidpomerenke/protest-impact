```python
%load_ext autoreload
%autoreload complete
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from tqdm.notebook import tqdm

from protest_impact.util import cache, project_root
```

```python
from protest_impact.data.covid_restrictions import load_mobility, load_stringency

df = load_mobility()
```

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 10), sharex=True, sharey=True)
i = 0
for col in df.columns:
    if col == "date":
        continue
    # z-score standardization
    y = (df[col] - df[col].mean()) / df[col].std()
    ax[i // 2, i % 2].plot(df["date"], y)
    ax[i // 2, i % 2].set_title(col)
    ax[i // 2, i % 2].set_ylabel("Change in visitors")
    ax[i // 2, i % 2].set_xlabel("Date")
    ax[i // 2, i % 2].axhline(0, color="black", linestyle="-", linewidth=1)
    i += 1
fig.autofmt_xdate()
fig.tight_layout()
plt.show()
```

```python
import optuna
```

```python
def get_study_df(version):
    study = optuna.create_study(
        direction="minimize",
        study_name=f"hyperopt_v{version}",
        load_if_exists=True,
        storage="sqlite:///../../hyperopt/synth.db",
    )
    df = study.trials_dataframe()
    df = df[df["state"] == "COMPLETE"]
    # remove the prefix "user_attrs_" from the column names
    df.columns = [c.replace("user_attrs_", "") for c in df.columns]
    df.columns = [c.replace("params_", "") for c in df.columns]
    return df


df = get_study_df(0.23)
```

The logquotient is a bad choice:

- For good estimators like linear regression, the mean is near 0 and the std is low anyway. Using the log ratio with a small smoothing constant skews the data in a way that introduces negative bias. Only using a larger smoothing constant reduces this effect, but there is little reason to justify it, and it will just make all positive and negative effects smaller. If a model benefits from using the log-ratio, then it is because its variance is very high or it has outliers, and I don't want either. So I should just look at the absolute difference, and also focus more on the variance (and perhaps outliers) to weed out bad models.

(However, the protest impacts may have a higher variance and outliers, so for them it might make sense to look at the log ratio. And then I might need the log ratio of random events for comparison. Not so sure about this.)

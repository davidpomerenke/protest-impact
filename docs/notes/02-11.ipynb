```python
import json

from protest_impact.util import project_root

with open(
    project_root
    / "data"
    / "protest"
    / "protest_news_shuffled_v2_sample_with_predictions.jsonl"
) as f:
    sample = [json.loads(line) for line in f]
```

```python
len([x for x in sample if x["label"] == 1])
```

```python
with open(
    project_root
    / "data"
    / "protest"
    / "protest_news_shuffled_v2_sample_with_positive_predictions.jsonl",
    "w",
) as f:
    for line in sample:
        if line["label"] == 1:
            f.write(json.dumps(line) + "\n")
```

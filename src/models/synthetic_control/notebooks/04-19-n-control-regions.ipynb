```python
from protest_impact.data.protests import get_climate_protests

climate_protests = get_climate_protests()
```

```python
from protest_impact.synthetic_region import get_regional_counts_for_protest
from tqdm.auto import tqdm

data = []
for protest in tqdm(climate_protests.to_dict("records")):
    df = get_regional_counts_for_protest(
        "klima*",
        protest,
        climate_protests,
        n_days_protest_free_pre=1,
        n_days_protest_free_post=1,
        min_control_regions=0,
        min_count=3,
    )
    length = len(df["region"].unique()) if df is not None else 0
    data.append((protest, length))
```

```python
counts = [
    dict(date=protest["event_date"], size=protest["size"], count=l)
    for protest, l in data
]
```

```python
import pandas as pd

df = pd.DataFrame(counts)
df
```

```python
# plot with date on x axis, counts on y axis, size as point size
import altair as alt

alt.Chart(df).mark_point().encode(
    x="date",
    y="count",
    size="size",
)
```

```python
import optuna


def objective(trial):
    param_a = trial.suggest_categorical("param_a", [1, 2])
    param_b = trial.suggest_categorical("param_b", [3, 4])
    param_c = trial.suggest_categorical("param_c", [5, 6])
    return 0


search_space = dict(
    param_a=[1, 2],
    param_b=[3, 4],
    param_c=[5, 6],
)

optuna.logging.set_verbosity(optuna.logging.WARNING)

for seed in range(100):
    study = optuna.create_study(
        sampler=optuna.samplers.GridSampler(search_space, seed=seed)
    )

    study.optimize(objective, n_trials=8, n_jobs=8)
    n_unique_params = len({tuple(s.params.items()) for s in study.trials})
    if n_unique_params != 8:
        print(f"{seed=} {n_unique_params=}")
        # print([s.params for s in study.trials])
```

```python
def objective(trial):
    param_a = trial.suggest_categorical("param_a", [1, 2])
    param_b = trial.suggest_categorical("param_b", [3, 4])
    if param_a == 2 and param_b == 4:
        param_c = trial.suggest_categorical("param_c", [5, 6])
    return 0


for seed in range(100):
    study = optuna.create_study(sampler=optuna.samplers.BruteForceSampler(seed=seed))

    study.optimize(objective, n_trials=8, n_jobs=8)
    n_unique_params = len({tuple(s.params.items()) for s in study.trials})
    if n_unique_params != 8:
        print(f"{seed=} {n_unique_params=}")
        # print([s.params for s in study.trials])
```
